point excition

```python
import numpy as np
from numpy import diag, ones, zeros, block
from scipy.linalg import block_diag as blkdiag
from scipy.linalg import eig
import matplotlib.pyplot as plt
N = 100
kappa = 1
H1=diag(kappa*ones((N-1)),1)+diag(kappa*ones((N-1)),-1)
H=zeros((0,0));
for i in range(N):
    H=blkdiag(H,H1);
H=H+kappa*diag(ones((N**2-N)),N)+kappa*diag(ones((N**2-N)),-N)
y0=zeros(N**2)
y0[round(N**2/2-N/2)]=1
fig, ax = plt.subplots()
ax.imshow(abs(y0.reshape(N,N))**2)
```

```python
z = 20
W, V = eig(H)
index = np.argsort(W)
W = W[index]
V = V[:,index]
y2 = V.dot((V.T.dot(y0))*np.exp(1j*W*z))
fig, ax = plt.subplots()
ax.imshow(abs(y2.reshape(N,N))**2)
```

```python
import numpy as np
from numpy import diag, ones, zeros, block
from scipy.linalg import block_diag as blkdiag
from scipy.linalg import eig
import matplotlib.pyplot as plt
N = 100
kappa = 1
H1=diag(kappa*ones((N-1)),1)+diag(kappa*ones((N-1)),-1)
H=zeros((0,0));
for i in range(N):
    H=blkdiag(H,H1);
H=H+kappa*diag(ones((N**2-N)),N)+kappa*diag(ones((N**2-N)),-N)
XX, YY = np.meshgrid(np.array(range(N)),np.array(range(N)))
y0 = np.exp(-((XX-50)**2+(YY-50)**2)/5**2).reshape(N**2)
fig, ax = plt.subplots()
ax.imshow(abs(y0.reshape(N,N))**2)
```
